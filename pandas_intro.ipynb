Группируем данные по gender и считаем среднее значение по writing score

```python
students_perf.groupby('gender').aggregate({'writing score': 'mean'})
```

Отберем первые три строки и первые 5 столбцов при помощи iloc

```python
students_perf.iloc[[1,2], 0:5]
```

```python
students_perf_with_names
```

Здесь мы можем отобрать данные не по индексам, а по их названиям.

```python
students_perf_with_names.loc[['Cersei', 'Tywin'], ['math score']]
```

Каждя колонка это пандовская серия

```python
pd.Series([1,2,3], index=['Первый', 'Второй', 'Третий'])
```

```python
my_series_1 = pd.Series([1,2,3], index=["first", 'second', 'third'])
my_series_2 = pd.Series([4,5,6], index=["first", 'second', 'third'])
```

из серий мы можем создавать датафрейм. По сути каждая колонка это серия.

```python
pd.DataFrame({'первая колонка': my_series_1, 'вторая колнка' : my_series_2})
```
